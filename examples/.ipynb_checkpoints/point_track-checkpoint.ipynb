 ## Point track method examples

```python
import turtle
import superTurtle as st
from svg_turtle import SvgTurtle
import matplotlib.pyplot as plt
import numpy as np
```

Let's draw deltoids following the trajectory of a sinusoidal function.

```python
def rotate(origin, point, angle):
    """
    Rotate a point counterclockwise by a given angle around a given origin.
    The angle should be given in degrees.
    """
    ox, oy = origin[0], origin[1]
    px, py = point[0], point[1]
    
    angle *= np.pi / 180.

    qx = ox + np.cos(angle) * (px - ox) - np.sin(angle) * (py - oy)
    qy = oy + np.sin(angle) * (px - ox) + np.cos(angle) * (py - oy)
    return [qx, qy]
```

```python
x = np.linspace(0., np.pi * 2., 60)
y = np.sin(x)
points = np.zeros((len(x), 2))

points[:, 0] = x
points[:, 1] = y

for i in range(len(points[:,0])):
    points[i,:] = rotate([0,0], points[i,:], 90)
    
plt.plot(points[:,0], points[:,1])
```

```python
turtle.screensize(1000, 1000)
t = turtle.Turtle()
t.speed(0)
#t = SvgTurtle(500, 500)
t.left(90)
# point to mark the center
t.dot(10)

delt =  st.SuperTurtle(t)
delt.tipo = "deltoid"
delt.colorRelleno = "blue"
delt.f = 0.7
delt.l = 18
delt.c = 12

sin =  st.SuperTurtle(t)
sin.tipo = "pointtrack"
sin.points = points
sin.scalepx = 400
sin.scalepy = 200
sin.penup = True
sin.center = True
sin.objd = (delt,)
sin.objdp = 0.5
sin.drawornot = [False] + [True, True, True, False] * len(points[:,0])

sin.draw()

turtle.exitonclick()

#dname = "point_track_1.svg"
#t.save_as(dname)
```

Let's draw deltoids around a circle

```python
# theta goes from 0 to 2pi
npuntos = 40
theta = np.linspace(0, 2*np.pi, npuntos)

# the radius of the circle
r = 1.

# compute x1 and x2
x1 = r*np.cos(theta)
x2 = r*np.sin(theta)

# create the figure
plt.plot(x1, x2, "b-")
for i in range(len(x1)):
    plt.text(x1[i], x2[i], str(i))
plt.show()
```

```python
points = np.array([x1, x2])
points = points.transpose()
points[:,0] -= 2 * r
points[:,1] -= 2 * r
```

```python
def modifyp(t, i, obj, objd, argsf = None):
    pal = argsf
    colp = st.colorpalette(pal)
    col = st.rgb_to_hex(colp[int(i) % len(colp)])
    objd[0].colorRelleno = "#" + col
    return 
```

```python
turtle.screensize(500, 500)
turtle.Screen().bgcolor("black")

t = turtle.Turtle()
#t = SvgTurtle(500, 500)
t.speed(0)
t.left(90)

t.dot(10)

delt = st.SuperTurtle(t)
delt.tipo = "deltoid"
delt.colorRelleno = "blue"
delt.f = 0.6666666
delt.l = 30
delt.c = 15

sin = st.SuperTurtle(t)
sin.tipo = "pointtrack"
sin.points = points
sin.scalepx = 300
sin.scalepy = 300
sin.penup = True
sin.center = True
sin.objd = (delt,)
sin.objdp = 0.5
sin.drawornot = [False] + ([True] + [True]) * len(points[:,0])
sin.argsf = "Pokemon"
sin.fmod = modifyp

sin.draw()

turtle.exitonclick()

#dname = "point_track_2.svg"
#t.save_as(dname)
```
